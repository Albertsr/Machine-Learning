```python
import numpy as np
import roc, prc, ks_curve
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
%matplotlib inline

X, y = make_classification(n_samples=10000, n_classes=2, n_features=20, 
                           n_informative=18, weights=[0.8, 0.2], random_state=2018)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2018)

clf = XGBClassifier()
clf.fit(X_train, y_train)
y_prob = clf.predict_proba(X_test)[:,-1]
```

#### ROC

```python
roc.plot_roc(y_test, y_prob)
```

#### PR_Curve

```python
prc.plot_prc(y_test, y_prob)
```

#### KS曲线

```python
ks_curve.plot_ks(y_test, y_prob)
```
